```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

```python
def handlerate(value):
    value = str(value).split('/')
    value = value[0];
    return float(value)
dataframe['rate'] = dataframe['rate'].apply(handlerate)
print(dataframe.head())
```

```python
dataframe.info()
```

```python
dataframe.head()
```

```python
sns.countplot(x= dataframe['listed_in(type)'])
plt.xlabel("type od restaurant")
              
```

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result,c='green', marker='o')
plt.xlabel("type of restaurant",c='red',size=20)
plt.ylabel("votes",c='red',size=20)
```

Conclusion: Dining type restraunts received maximum votes

```python
plt.hist(dataframe['rate'],bins=5)
plt.title("Rate of restaurant")
plt.show()
```

Conclusion: The majority rating of restraunts lies between 3.5 to 4

```python
couple_data = dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

Conclusion: The majority of couples preferred order of approximate cost of 300rs

```python
plt.figure(figsize =(6,6))
sns.boxplot(x='online_order',y ='rate', data=dataframe)
```

Conclusion:
 Offline prders received lower rating compared to onlune orders

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc= 'size', fill_value=0)
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Heatmap')
plt.xlabel('Online Order')
plt.ylabel('Listed In (Type)')
plt.show()
```

Conclusion: dining restraunts primarly have offilne orders, whereas cafes primarly receive online orders
