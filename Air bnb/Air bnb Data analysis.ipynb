```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('D:\compressed_data (1).csv.gz', compression='gzip')
```

```python
df.head()
```

```python
df.columns
```

```python
print(df.isnull().sum())
```

```python
df.info()
```

```python
df['last review']= pd.to_datetime(df['last review'], errors= 'coerce')

```

```python
df.head()
```

```python
df.fillna({'reviews per month': 0, 'last review' : df['last review'].min()}, inplace=True)
```

```python
df.dropna(subset=['NAME', 'host name'], inplace= True)
```

```python
print(df.isnull().sum())
```

```python
df = df.drop(columns= ["license", "house_rules"], errors ='ignore')
```

```python
df.head()
```

```python
df['price']= df['price'].replace('[\$,]', '', regex=True).astype(float)
df['service fee']= df['service fee'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True, color="red")
plt.title('Distribution of listing price')
plt.xlabel('Price $')
plt.ylabel('Frequency')
plt.show()
```

Conclusion: This histogram shows evenly distributed listing price across variety

```python
df['room type']
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='room type', data=df, color='pink')
plt.title('Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.show()
```

Conclusion: Most people book entire home/apt and least shared room

```python
plt.figure(figsize=(10, 6))
sns.countplot(y='neighbourhood group', data=df, color='lightgreen',order=df['neighbourhood group'].value_counts().index)
plt.title('Neighbourhood Group Distribution')
plt.xlabel('Count')
plt.ylabel('Neighbourhood Group')
plt.show()
```

Conclusion: Highest listing is done from Manhattan

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='room type', y='price', data=df, palette='Set1', hue='room type')
plt.title('Price vs Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price $')
plt.legend(title='Room Type')
plt.show()
```

Conclusion: This boxplot shows relationship between Room type and price

```python
df['last review']= pd.to_datetime(df['last review'], errors= 'coerce')
reviews_over_time = df.groupby(df['last review'].dt.to_period('M')).size()
plt.figure(figsize=(12, 6))
reviews_over_time.plot(kind='line', marker='o', color='blue')
plt.title('Number of Reviews Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Reviews')
plt.show()
```

Conclusion: This graph shows that reviews increased from 2019 tp 2024
